Numpy Arrays

```python
#numpy arrays:
np_array=np.array([1,2,3,4])
print(np_array)
type(np_array) #ndarray--> nth dimensional array,similiar to matrix
```

```python
# 1D array:
a=np.array([1,2,3,4])
print(a)
```

```python
#2D array:
b=np.array([(1,2,3,4),(5,6,7,8)])
print(b)
b.shape #--> shape= (rows, columns)
```

initial placeholers in numpy array:


â„¹ initialize (like int a; in c)


```python
# create a numpy array of zeros
import numpy as np
x = np.zeros((4,5)) #zeros() creates a array of 0's where there is 4 rows and 5 column.
print(x)
```

```python
# numpy array of 1's
y=np.ones((3,3))
print(y)
```

```python
# array of particular value:
z = np.full((5,4),5)
print(z)
```

```python
# identity matrix
i = np.eye(5, dtype=int)
print(i)
```

```python
# create a numpy array with random values:
r = np.random.random((3,4))
print(r)
```

```python
# random integer values array within a range.
r2=np.random.randint(10,50, (3,5))
print(r2)
```

```python
# array of evenly spaced values: specifying the number of values required.
e = np.linspace(10,30,8, dtype=int)
print(e)
```

```python
# array of evenly spaced values with step
e2 = np.arange(10,30,6)
print(e2)
```

```python
# convert the list to numpy array
list2=[10,20,30,40,50]

np_array=np.asarray(list2)
print(np_array)
type(np_array)
```

Array manipulation

```python
# array manipulation.
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#transpose:
trans= np.transpose(array)
print(trans)
print(trans.shape)
```

```python
trans2= array.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
