Matplot lib:

Useful for plotting graphs.

```python
# Importing Matplotlib lirbrary
import matplotlib.pyplot as plt
```

```python
# importing numpy library to get data to plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data.

```python
# sine wave:
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.show()
```

```python
# plot a parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('x------->')
plt.ylabel('y=x^2')
plt.title('Parabola curve')
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure() #empty plot
ax = fig.add_axes([0,0,1,1]) #(0,0) origin/coordinates and 1,1 is height-width of rectangle
languages = ['english', 'French', 'Spanish', 'Latin', 'German']
people = [100,50,150,40,65]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart:

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['english', 'French', 'Spanish', 'Latin', 'German']
people = [100,50,150,40,65]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scattered Plot

good for clusturing.

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
